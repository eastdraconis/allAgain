## 2020년도 쓰레기에 대한 환경오염 방지 노력
  1. 적극적인 노력에 대한 평균

```python
import numpy as np 
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt
import math
%matplotlib inline

## 한글 폰트 적용 안되는 것에 대한 코드
matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False

## 각 데이터셋 가져오기
df1 = pd.read_csv('환경오염_방지_노력__음식물_쓰레기_줄이기__13세_이상_인구.csv', encoding="CP949")
df2 = pd.read_csv('환경오염_방지_노력_가정_내_대기전력_줄이기__13세_이상_인구.csv', encoding="CP949")
df3 = pd.read_csv('환경오염_방지_노력_대중교통_이용하기__13세_이상_인구.csv', encoding="CP949")
df4 = pd.read_csv('환경오염_방지_노력_물_절약하기__13세_이상_인구.csv', encoding="CP949")
df5 = pd.read_csv('환경오염_방지_노력_일회용품_사용하지_않기__13세_이상_인구.csv', encoding="CP949")
df6 = pd.read_csv('환경오염_방지_노력_자연보호_및_환경보전활동_참여하기__13세_이상_인구.csv', encoding="CP949")
df7 = pd.read_csv('환경오염_방지_노력_재활용품_분리_배출하기__13세_이상_인구.csv', encoding="CP949")
df8 = pd.read_csv('환경오염_방지_노력_친환경제품_구입·사용하기__13세_이상_인구.csv', encoding="CP949")
df9 = pd.read_csv('환경오염_방지_노력_합성세제_사용_줄이기__13세_이상_인구.csv', encoding="CP949")

## 각각의 데이터셋에서 총계만 뽑기
df1Effort = df1.iloc[1]["2020.1"]
df2Effort = df2.iloc[1]["2020.1"]
df3Effort = df3.iloc[1]["2020.1"]
df4Effort = df4.iloc[1]["2020.1"]
df5Effort = df5.iloc[1]["2020.1"]
df6Effort = df6.iloc[1]["2020.1"]
df7Effort = df7.iloc[1]["2020.1"]
df8Effort = df8.iloc[1]["2020.1"]
df9Effort = df9.iloc[1]["2020.1"]


## 하나의 array로 만들어주기
effortScore = [df1Effort,df2Effort,df3Effort,df4Effort,df5Effort,df6Effort,df7Effort,df8Effort,df9Effort]
effortList = ["음식물쓰레기줄이기","가정내대기전력줄이기","대중교통이용하기","물절약하기","일회용품사용하지않기","자연보호및환경보전활동참여하기","재활용품분리배출하기","친환경제품구입·사용하기","합성세제사용줄이기"]

## numpy.array type으로 변경해주기
x = np.array(effortList)
## numpy.array type으로 변경해주면서 내용의 type은 float64로 변경
y = np.array(effortScore).astype(np.float64)

## 평균값
total_avg = math.ceil(y.mean())


fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.25
plt.bar(x,y,bar_width, alpha=0.4, color='orange', label='매우노력함')
plt.axhline(y=total_avg,xmax=1,color="green",linestyle="dashed",label='평균')
plt.xticks(rotation=45)
plt.title("환경오염 방지 노력 부문")
plt.xlabel("부문")
plt.ylabel("수치")
plt.legend()
plt.show()



```
