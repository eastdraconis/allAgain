```python


## 하나의 array로 만들어주기
effortScore1 = [df1Effort1,df2Effort1,df3Effort1,df4Effort1,df5Effort1,df6Effort1,df7Effort1,df8Effort1,df9Effort1]
effortScore2 = [df1Effort2,df2Effort2,df3Effort2,df4Effort2,df5Effort2,df6Effort2,df7Effort2,df8Effort2,df9Effort2]
effortScore3 = [df1Effort3,df2Effort3,df3Effort3,df4Effort3,df5Effort3,df6Effort3,df7Effort3,df8Effort3,df9Effort3]
effortScore4 = [df1Effort4,df2Effort4,df3Effort4,df4Effort4,df5Effort4,df6Effort4,df7Effort4,df8Effort4,df9Effort4]
effortList = ["음식물쓰레기줄이기","가정내대기전력줄이기","대중교통이용하기","물절약하기","일회용품사용하지않기","자연보호및환경보전활동참여하기","재활용품분리배출하기","친환경제품구입·사용하기","합성세제사용줄이기"]
```

```python

## numpy.array type으로 변경해주기
index = np.arange(9)
x = np.array(effortList)
## numpy.array type으로 변경해주면서 내용의 type은 float64로 변경
y1 = np.array(effortScore1).astype(np.float64)
y2 = np.array(effortScore2).astype(np.float64)
y3 = np.array(effortScore3).astype(np.float64)
y4 = np.array(effortScore4).astype(np.float64)

## 매우노력함평균값
total_avg = math.ceil(y1.mean())

```

```python

fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.2
ax.bar(index,y1,bar_width, alpha=0.4, color='blue', label='매우 노력함')
ax.bar(index+bar_width,y2,bar_width, alpha=0.4, color='green', label='약간 노력함')
ax.bar(index+(bar_width*2),y3,bar_width, alpha=0.4, color='orange', label='별로 노력하지 않음')
ax.bar(index+(bar_width*3),y4,bar_width, alpha=0.4, color='red', label='전혀 노력하지 않음')
# ax.axhline(y=total_avg,xmax=1,color="green",linestyle="dashed",label='평균')
ax.set_xticks(np.arange(bar_width, 9 + bar_width, 1), x,rotation=45)
ax.set_title("2020년 환경오염 방지 노력에 대한 설문")
ax.set_xlabel("부문")
ax.set_ylabel("%")
fig.legend()
plt.show()
```
